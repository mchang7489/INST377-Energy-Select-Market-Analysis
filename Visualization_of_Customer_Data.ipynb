```python
# Import Packages

import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
# Import Customer Data 

df = pd.read_csv('/content/Customer.Data.csv', encoding= 'unicode_escape')

# Display Customer Data
df
```

```python
# Status of Customer Data

Status = df[['Status','Source']].groupby(['Status']).agg('count')

Status = Status.reset_index()
Status.columns = ["Status","Count"]
Status
```

## Reviewing Customer Data

```python
# Different Sources used in Energy Select LLC

source = df[['City','Source']].groupby(['Source']).agg('count')

source = source.reset_index()
source.columns = ["Source","Count"]
source
```

## Visualizations

```python
# The Number of Cities

City = df['City'].value_counts().plot(kind='bar',
                                    figsize=(14,8),
                                    title="The Number of Cities")
City.set_xlabel("City")
City.set_ylabel("Frequency")
```

```python
# The Number of Sources

Source_Type = df['Source'].value_counts().plot(kind='pie',
                                    figsize=(14,8),
                                    title="The Number of Sources")
Source_Type.set_xlabel("Source")
Source_Type.set_ylabel("Frequency")
```

```python
# Status of each Customers

Status.groupby(['Status']).sum().plot(kind='pie', y='Count', autopct='%1.0f%%',
                                colors = ['red', 'pink', 'steelblue'],
                                title='Status of each Customers')
```
